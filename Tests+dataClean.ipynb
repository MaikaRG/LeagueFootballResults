```python
import requests
from bs4 import BeautifulSoup

newTable=requests.get('https://fbref.com/en/comps/12/1886/schedule/2018-2019-La-Liga-Fixtures')
htmlNew= newTable.text
soupNew = BeautifulSoup(htmlNew, 'html.parser')
soupNew
```

```python
filas = soupNew.select("#sched_ks_1886_1 > tbody:nth-child(4)")
x = filas[0].find_all("a")
date=[]
local=[]
visitante=[]
link=[]
import re
for e in range(0,len(x),5):
    date.append(x[e].text)
    local.append(x[e+1].text)
    visitante.append(x[e+3].text)
    link.append(x[e+4])
print(len(date))
print(len(local))
print(len(visitante))
links=[]
for e in link:
    e =e.get('href')
    h = ("https://fbref.com"+e)
    links.append(h)
#print((links))
print(links)
```

```python
print(visitante)
```

```python
import pandas as pd
df = pd.DataFrame(data = {"Date":date, "Local":local, "Visitor":visitante, "Links":links})
display(df)
df.dtypes
```

```python
temporada1819 = pd.read_csv('./Inputs/season-1819.csv')
temporada1819.head()
temporada1819['Date'] = pd.to_datetime(temporada1819['Date'])
temporada1819.head()
temporada1819=temporada1819[['HomeTeam', 'AwayTeam','FTHG','FTAG']]
temporada1819.columns = ['Local','Visitor',
                     'GoalsLocal','GoalsVisitor']
temporada1819.head()
temporada1819.dtypes
#temporada1819["Date"] = temporada1819["Date"].astype(str)
temporada1819
```

```python
print(set(temporada1819['Local']))
print(set(df['Local']))

def team_fixer(text):
    team_dict = {
        'Sociedad':'Real Sociedad',
        'Levante':'Levante',
        'Barcelona':'Barcelona',
        'Valencia':'Valencia',
        'Real Madrid':'Real Madrid',
        'Vallecano':'Rayo Vallecano',
        'Girona':'Girona',
        'Eibar':'Eibar',
        'Betis':'Betis',
        'Leganes':'Leganés',
        'Alaves':'Alavés',
        'Ath Madrid':'Atlético Madrid',
        'Valladolid':'Valladolid',
        'Espanol':'Espanyol',
        'Sevilla':'Sevilla',
        'Huesca':'Huesca',
        'Celta':'Celta Vigo',
        'Getafe':'Getafe',
        'Villarreal':'Villarreal',
        'Ath Bilbao':'Athletic Bilbao'
    }
    if text in team_dict:
        return team_dict[text]

temporada1819['Local']=temporada1819['Local'].apply(team_fixer)
temporada1819['Visitor']=temporada1819['Visitor'].apply(team_fixer)
```

```python
df_final = pd.merge(df, temporada1819, on=['Local','Visitor'])
df_final=df_final[['Date','Local','GoalsLocal','GoalsVisitor','Visitor','Links']]
```

```python
import re
def resultado(local,visitante):
    df_consulta = df_final[df_final["Local"]==local]
    df_consulta = df_consulta[df_final["Visitor"]==visitante]
    res = requests.get(list(df_consulta["Links"])[0])
    html = res.text
    soup = BeautifulSoup(html, 'html.parser')
    golesPaginaB = soup.select('.scorebox > div:nth-child(5)')
    listaGoalCard=[]
    for e in golesPaginaB:
        e1 = e.text
        e1 = (e1.strip().split('\n'))
        e2 = e.select('.event_icon')
        for e in e2:
            e = str(e)
            listaGoalCard.append(e)
    visitorEvents=list(zip(listaGoalCard,e1))
    listGolesVisitante=[]
    for e in visitorEvents:
        for i in e:
            if re.search("goal",i):
                a=''.join(e[1])
                a = a.split('·')
                b = 'en el minuto'.join(a)
                listGolesVisitante.append(re.sub('&rsquor',"'",b))
    golesPaginaA = soup.select('.scorebox > div:nth-child(4)')
    listaGoalCard=[]
    for e in golesPaginaA:
        e1 = e.text
        e1 = (e1.strip().split('\n'))
        e2 = e.select('.event_icon')
        for e in e2:
            e = str(e)
            listaGoalCard.append(e)
    localEvents=list(zip(listaGoalCard,e1))
    listGolesLocal=[]
    for e in localEvents:
        for i in e:
            if re.search("goal",i):
                a=''.join(e[1])
                a = a.split('·')
                b = 'en el minuto'.join(a)
                listGolesLocal.append(re.sub('&rsquor',"'",b))
    pintaGoles(local,listGolesLocal)
    pintaGoles(visitante, listGolesVisitante)
```

```python
def pintarMediaGolesEnSuRol(local,visitante):
    df_consulta = df_final[df_final["Local"]==local]
    df_consultaVis = df_final[df_final["Visitor"]==local]
    a =(df_consulta['GoalsLocal'].mean())
    b =(df_consultaVis['GoalsVisitor'].mean())
    #display(df_consulta)
    #display(df_consultaVis)
    df_consulta2 = df_final[df_final["Local"]==visitante]
    df_consultaVis2 = df_final[df_final["Visitor"]==visitante]
    c =(df_consulta2['GoalsLocal'].mean())
    d =(df_consultaVis2['GoalsVisitor'].mean())
    #display(df_consulta2)
    #display(df_consultaVis2)
    x=(local+" loc.",visitante+" loc.",local+" vis.",visitante+" vis.")
    y=(a,c,b,d)
    plt.bar(x,y)
    plt.title('Media de goles en su rol(local/visitante)', fontsize=17, color='Blue')
    plt.xlabel("Equipos")
    plt.ylabel("Media")
    plt.savefig('test.png')
```

```python
def pintarMaxGolesEnSuRol(local,visitante):
    df_consulta = df_final[df_final["Local"]==local]
    df_consultaVis = df_final[df_final["Visitor"]==local]
    a =(df_consulta['GoalsLocal'].max())
    b =(df_consultaVis['GoalsVisitor'].max())
    #display(df_consulta)
    #display(df_consultaVis)
    df_consulta2 = df_final[df_final["Local"]==visitante]
    df_consultaVis2 = df_final[df_final["Visitor"]==visitante]
    c =(df_consulta2['GoalsLocal'].max())
    d =(df_consultaVis2['GoalsVisitor'].max())
    #display(df_consulta2)
    #display(df_consultaVis2)
    x=(local+" loc.",visitante+" loc.",local+" vis.",visitante+" vis.")
    y=(a,c,b,d)
    plt.bar(x,y)
    plt.title('Máximo número de goles en su rol(local/visitante)', fontsize=17, color='Blue')
    plt.xlabel("Equipos")
    plt.ylabel("Media")
```
